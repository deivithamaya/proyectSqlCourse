```python
try:
    #ya mire que mi base de datos fue creada , prosigo a dejar los datos como los necesito
    mydb = mysql.connector.connect(
        host="localhost",
        user="root",
        password="12345",
        database="tareaSql"
    )
    for lisP,lisN in zip(listDf,listdi):
        names = lisP.columns.values
        types = lisP.dtypes
        name = lisN.split(".")[0]
        listOfState = []
        statement = "create table if not exists " + name +"("
        for i,j in zip(names,types):
            if j =="int64":
                statement += "`"+i+"`"+" integer" 
            elif j =="float64":
                statement += "`"+i+"`"+" decimal(15,5)"
            else :
                statement += "`"+i+"`"+" varchar(100)"
            if i == names[-1]:statement+=");"
            else:
                statement+=","
        print(statement)
        with mydb.cursor() as cur:
            cur.execute(statement)
            sta = "show tables;"
            cur.execute(sta)
            resul = cur.fetchall()
            print(resul)
        
        engine = create_engine("mysql+mysqlconnector://root:12345@localhost/tareaSql")
        lisP.to_sql(name=name,con=engine, if_exists = 'append',index = False)
        
except Exception as e:
    print("ha ocurrido un error:")
    print(e)
```
